```python
df_benign = pd.concat([
    pd.read_csv("stateless_features-benign_1.pcap.csv"),
    pd.read_csv("stateless_features-benign_2.pcap.csv"),
], ignore_index=True)

print(df_benign.shape)
print(df_benign.describe())
```

```python
features = [
    "FQDN_count",
    "subdomain_length",
    "upper",
    "lower",
    "numeric",
    "special",
    "labels",
    "labels_max",
    "labels_average",
    "len",
    "subdomain",
]
```

```python
X_benign = df_benign[features].values
print(X_benign.shape)
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(
    n_estimators = 100,
    contamination="auto",
    random_state=42,
    n_jobs=-1,
)
iso.fit(X_benign)
```

```python
import glob

attack_files = glob.glob("Attacks/stateless_features-light_*.pcap.csv") + \
                glob.glob("Attacks-2/stateless_features-heavy_*.pcap.csv")

print(f"Found {len(attack_files)} attack files:")
for f in attack_files:
    print(f"  {f}")
```

```python
df_attacks = pd.concat([
    pd.read_csv(f) for f in attack_files
], ignore_index=True)
print(df_attacks.shape)
print(df_attacks.columns.tolist())
```

```python
import numpy as np

scores_benign = iso.decision_function(df_benign[features].values)
scores_attacks = iso.decision_function(df_attacks[features].values)

print(f"Benign - mean: {scores_benign.mean():.4f}, std: {scores_benign.std():.4f}")
print(f"Attack - mean: {scores_attacks.mean():.4f}, std: {scores_attacks.std():.4f}")
print(f"\nBenign  percentiles: 1%={np.percentile(scores_benign, 1):.4f}, 5%={np.percentile(scores_benign, 5):.4f}, 50%={np.percentile(scores_benign, 50):.4f}")
print(f"Attack  percentiles: 50%={np.percentile(scores_attacks, 50):.4f}, 95%={np.percentile(scores_attacks, 95):.4f}, 99%={np.percentile(scores_attacks, 99):.4f}")
```
